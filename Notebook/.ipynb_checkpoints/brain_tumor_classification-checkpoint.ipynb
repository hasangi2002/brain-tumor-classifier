- Loads real brain MRI images (4 classes)
- Trains a custom CNN built from scratch
- Trains ResNet18 using Transfer Learning
- Compares both models on test data

**Classes:** Glioma · Meningioma · Pituitary · No Tumor

## Step 3 — Data Preprocessing & Augmentation

```python
# ── Image Size & Batch Size ───────────────────────────
IMG_SIZE   = 224    # resize all images to 224x224 pixels
BATCH_SIZE = 32     # process 32 images at a time during training

# ── Training Transforms ───────────────────────────────
# Augmentation = artificially create variety so model doesn't memorise
train_transforms = transforms.Compose([
    transforms.Resize((IMG_SIZE, IMG_SIZE)),      # resize to 224x224
    transforms.RandomHorizontalFlip(p=0.5),       # 50% chance flip left/right
    transforms.RandomRotation(degrees=15),        # rotate up to 15 degrees
    transforms.ColorJitter(brightness=0.2,        # slight brightness change
                           contrast=0.2),         # slight contrast change
    transforms.ToTensor(),                        # convert image to numbers (0-1)
    transforms.Normalize(                         # normalise using ImageNet stats
        mean=[0.485, 0.456, 0.406],               # average colour values
        std=[0.229, 0.224, 0.225]                 # spread of colour values
    )
])

# ── Validation/Test Transforms ────────────────────────
# No augmentation here — we want real unmodified images for testing
val_transforms = transforms.Compose([
    transforms.Resize((IMG_SIZE, IMG_SIZE)),
    transforms.ToTensor(),
    transforms.Normalize(
        mean=[0.485, 0.456, 0.406],
        std=[0.229, 0.224, 0.225]
    )
])

print("Transforms defined!")
print(f"  Image size  : {IMG_SIZE}x{IMG_SIZE}")
print(f"  Batch size  : {BATCH_SIZE}")
```

```python
# ── Load Images from Folders ──────────────────────────
# ImageFolder automatically reads subfolders as class labels
full_train   = datasets.ImageFolder(DATA_DIR, transform=train_transforms)
test_dataset = datasets.ImageFolder(TEST_DIR,  transform=val_transforms)

# ── Split Training into Train (80%) + Validation (20%) ─
train_size = int(0.8 * len(full_train))   # 80% for training
val_size   = len(full_train) - train_size  # 20% for validation

train_dataset, val_dataset = random_split(
    full_train, 
    [train_size, val_size],
    generator=torch.Generator().manual_seed(42)  # reproducible split
)

# ── DataLoaders ───────────────────────────────────────
# DataLoader feeds images to the model in batches automatically
train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, 
                          shuffle=True)   # shuffle so model doesn't learn order
val_loader   = DataLoader(val_dataset,   batch_size=BATCH_SIZE, 
                          shuffle=False)  # no shuffle for validation
test_loader  = DataLoader(test_dataset,  batch_size=BATCH_SIZE, 
                          shuffle=False)  # no shuffle for testing

# ── Summary ───────────────────────────────────────────
print("Dataset loaded successfully!")
print(f"  Train      : {len(train_dataset):,} images")
print(f"  Validation : {len(val_dataset):,} images")
print(f"  Test       : {len(test_dataset):,} images")
print(f"  Classes    : {full_train.classes}")
```

## Step 4 — Build Custom CNN from Scratch

### How our CNN works:
- **Block 1–4:** Conv2d → BatchNorm → ReLU → MaxPool
- Each block **doubles the filters** (32 → 64 → 128 → 256)
- Each MaxPool **halves the image size** (224 → 112 → 56 → 28 → 14)
- **Classifier:** Flatten → FC(512) → Dropout → Output(4 classes)

```python
class BrainTumorCNN(nn.Module):
    """Custom CNN for Brain Tumor Classification."""

    def __init__(self, num_classes=4):
        super(BrainTumorCNN, self).__init__()

        # ── Feature Extractor ─────────────────────────
        # Learns WHAT to look for in the image (edges, shapes, tumors)
        self.features = nn.Sequential(

            # Block 1 — finds basic edges and shapes
            # Input: 3 colour channels, Output: 32 feature maps
            nn.Conv2d(3, 32, kernel_size=3, padding=1),
            nn.BatchNorm2d(32),    # stabilises training
            nn.ReLU(inplace=True), # activation — keeps only positive values
            nn.MaxPool2d(2, 2),    # shrinks image: 224 → 112

            # Block 2 — finds more complex patterns
            # Input: 32 feature maps, Output: 64 feature maps
            nn.Conv2d(32, 64, kernel_size=3, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(2, 2),    # shrinks image: 112 → 56

            # Block 3 — finds high-level features
            # Input: 64 feature maps, Output: 128 feature maps
            nn.Conv2d(64, 128, kernel_size=3, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(2, 2),    # shrinks image: 56 → 28

            # Block 4 — finds complex tumor structures
            # Input: 128 feature maps, Output: 256 feature maps
            nn.Conv2d(128, 256, kernel_size=3, padding=1),
            nn.BatchNorm2d(256),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(2, 2),    # shrinks image: 28 → 14
        )

        # ── Classifier ────────────────────────────────
        # Takes extracted features and decides the class
        self.classifier = nn.Sequential(
            nn.AdaptiveAvgPool2d((4, 4)),  # force output to 4x4 regardless of input
            nn.Flatten(),                  # convert 256x4x4 into a single list of 4096
            nn.Linear(256 * 4 * 4, 512),  # fully connected layer: 4096 → 512
            nn.ReLU(inplace=True),
            nn.Dropout(0.5),               # randomly turn off 50% of neurons
                                           # prevents memorising training data
            nn.Linear(512, num_classes)    # final output: 512 → 4 class scores
        )

    def forward(self, x):
        # This defines how data flows through the network
        x = self.features(x)      # extract features
        x = self.classifier(x)    # classify
        return x


# ── Create the model ──────────────────────────────────
cnn_model = BrainTumorCNN(num_classes=4).to(device)

# Count how many parameters (numbers) the model has
total_params = sum(p.numel() for p in cnn_model.parameters())
print(f"Custom CNN created!")
print(f"Total parameters: {total_params:,}")
print(f"Running on: {device}")
```

## Step 5 — Training Functions

```python
def train_one_epoch(model, loader, criterion, optimizer):
    """
    Train the model for ONE epoch (one full pass through training data).
    Returns: average loss, average accuracy
    """
    model.train()   # set model to training mode (enables dropout, batchnorm)
    
    total_loss = 0.0
    correct    = 0
    total      = 0

    for images, labels in loader:
        # Move data to GPU/CPU
        images = images.to(device)
        labels = labels.to(device)

        # ── Forward Pass ──────────────────────────────
        optimizer.zero_grad()        # clear old gradients
        outputs = model(images)      # get predictions
        loss    = criterion(outputs, labels)  # calculate how wrong we are

        # ── Backward Pass ─────────────────────────────
        loss.backward()    # calculate gradients (how to improve)
        optimizer.step()   # update model weights

        # ── Track statistics ──────────────────────────
        total_loss += loss.item() * images.size(0)
        _, predicted = outputs.max(1)             # get class with highest score
        correct      += predicted.eq(labels).sum().item()
        total        += labels.size(0)

    avg_loss = total_loss / total
    avg_acc  = correct / total
    return avg_loss, avg_acc


print("train_one_epoch() defined!")
```

```python
def evaluate(model, loader, criterion):
    """
    Evaluate the model on validation or test data.
    No learning happens here — just measuring performance.
    Returns: average loss, average accuracy
    """
    model.eval()   # set model to evaluation mode (disables dropout)

    total_loss = 0.0
    correct    = 0
    total      = 0

    with torch.no_grad():   # don't calculate gradients (saves memory)
        for images, labels in loader:
            images = images.to(device)
            labels = labels.to(device)

            outputs = model(images)
            loss    = criterion(outputs, labels)

            total_loss += loss.item() * images.size(0)
            _, predicted = outputs.max(1)
            correct      += predicted.eq(labels).sum().item()
            total        += labels.size(0)

    avg_loss = total_loss / total
    avg_acc  = correct / total
    return avg_loss, avg_acc


print("evaluate() defined!")
```

```python
def train_model(model, train_loader, val_loader, num_epochs=20, lr=0.001):
    """
    Full training loop — trains for multiple epochs and saves the best model.
    Returns: history (loss and accuracy per epoch)
    """

    # ── Loss & Optimizer ──────────────────────────────
    criterion = nn.CrossEntropyLoss()   # standard loss for classification
    optimizer = optim.Adam(             # Adam adjusts learning rate automatically
        model.parameters(), 
        lr=lr,              # learning rate — how big each update step is
        weight_decay=1e-4   # small penalty for large weights (reduces overfitting)
    )
    # Reduce learning rate every 7 epochs to fine-tune
    scheduler = optim.lr_scheduler.StepLR(optimizer, step_size=7, gamma=0.1)

    # ── History ───────────────────────────────────────
    # Store loss and accuracy for each epoch so we can plot later
    history = {
        'train_loss': [], 'val_loss': [],
        'train_acc':  [], 'val_acc':  []
    }

    best_val_acc = 0.0   # track best accuracy to save best model

    print(f"Starting training for {num_epochs} epochs...")
    print("=" * 65)

    for epoch in range(num_epochs):
        t0 = time.time()   # start timer

        # Train and evaluate
        train_loss, train_acc = train_one_epoch(model, train_loader, 
                                                criterion, optimizer)
        val_loss,   val_acc   = evaluate(model, val_loader, criterion)
        scheduler.step()   # update learning rate

        # Save to history
        history['train_loss'].append(train_loss)
        history['val_loss'].append(val_loss)
        history['train_acc'].append(train_acc)
        history['val_acc'].append(val_acc)

        # Save model if this is the best so far
        if val_acc > best_val_acc:
            best_val_acc = val_acc
            torch.save(model.state_dict(), 
                       f'../results/{model.__class__.__name__}_best.pth')
            saved = " ✓ saved"
        else:
            saved = ""

        elapsed = time.time() - t0
        print(f"Epoch [{epoch+1:02d}/{num_epochs}] "
              f"Train Loss: {train_loss:.4f} Acc: {train_acc*100:.1f}% | "
              f"Val Loss: {val_loss:.4f} Acc: {val_acc*100:.1f}% | "
              f"{elapsed:.1f}s{saved}")

    print("=" * 65)
    print(f"Training complete! Best val accuracy: {best_val_acc*100:.2f}%")
    return history


print("train_model() defined!")
```

## Step 7 — Transfer Learning with ResNet18

```python
def build_resnet18(num_classes=4):
    # Load ResNet18 pretrained on ImageNet (1.2 million images)
    model = models.resnet18(weights=models.ResNet18_Weights.IMAGENET1K_V1)

    # Freeze all layers — keep the knowledge it already has
    for param in model.parameters():
        param.requires_grad = False

    # Replace only the final layer for our 4 classes
    in_features = model.fc.in_features   # 512 neurons
    model.fc = nn.Sequential(
        nn.Linear(in_features, 256),
        nn.ReLU(inplace=True),
        nn.Dropout(0.4),
        nn.Linear(256, num_classes)      # 4 output classes
    )

    trainable = sum(p.numel() for p in model.parameters() 
                    if p.requires_grad)
    total     = sum(p.numel() for p in model.parameters())
    print(f"ResNet18 loaded!")
    print(f"Trainable parameters : {trainable:,}")
    print(f"Total parameters     : {total:,}")
    print(f"Frozen parameters    : {total - trainable:,}")
    return model


# Create the model
resnet_model = build_resnet18(num_classes=4).to(device)
```

```python
print("Training ResNet18 (Transfer Learning)...")
print("This will be faster and more accurate!\n")

resnet_history = train_model(
    model        = resnet_model,
    train_loader = train_loader,
    val_loader   = val_loader,
    num_epochs   = 5,
    lr           = 0.001
)
```

## Step 8 — Plot Training Curves

```python
def plot_history(history, title, color):
    epochs = range(1, len(history['train_loss']) + 1)
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(13, 5))

    # Loss plot
    ax1.plot(epochs, history['train_loss'], 
             color=color, linestyle='-', marker='o', markersize=5, label='Train Loss')
    ax1.plot(epochs, history['val_loss'],   
             color=color, linestyle='--', marker='s', markersize=5, label='Val Loss')
    ax1.set_title(f'{title} — Loss', fontweight='bold')
    ax1.set_xlabel('Epoch')
    ax1.set_ylabel('Loss')
    ax1.legend()
    ax1.grid(alpha=0.3)

    # Accuracy plot
    ax2.plot(epochs, [a*100 for a in history['train_acc']], 
             color=color, linestyle='-', marker='o', markersize=5, label='Train Acc')
    ax2.plot(epochs, [a*100 for a in history['val_acc']],   
             color=color, linestyle='--', marker='s', markersize=5, label='Val Acc')
    ax2.set_title(f'{title} — Accuracy', fontweight='bold')
    ax2.set_xlabel('Epoch')
    ax2.set_ylabel('Accuracy (%)')
    ax2.legend()
    ax2.grid(alpha=0.3)

    plt.tight_layout()
    plt.savefig(f'../results/{title.lower().replace(" ","_")}_curves.png', 
                dpi=150, bbox_inches='tight')
    plt.show()
    print(f"Saved!")

# Plot both models
plot_history(cnn_history,    'Custom CNN',          '#3498DB')
plot_history(resnet_history, 'ResNet18 Transfer',   '#E74C3C')
```

## Step 9 — Evaluate on Test Set

```python
def get_predictions(model, loader):
    """Run model on test set and collect all predictions."""
    model.eval()
    all_preds  = []
    all_labels = []

    with torch.no_grad():
        for images, labels in loader:
            images  = images.to(device)
            outputs = model(images)
            _, preds = outputs.max(1)
            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(labels.numpy())

    return np.array(all_labels), np.array(all_preds)


CLASS_NAMES = ['Glioma', 'Meningioma', 'No Tumor', 'Pituitary']

print("get_predictions() defined!")
```

```python
# Load best saved weights for Custom CNN
cnn_model.load_state_dict(
    torch.load('../results/BrainTumorCNN_best.pth', 
               map_location=device)
)

# Get predictions
cnn_labels, cnn_preds = get_predictions(cnn_model, test_loader)

# Calculate accuracy
cnn_test_acc = (cnn_labels == cnn_preds).mean() * 100

print(f"Custom CNN — Test Accuracy: {cnn_test_acc:.2f}%")
print("=" * 55)
print(classification_report(cnn_labels, cnn_preds, 
                             target_names=CLASS_NAMES))
```

```python
# Load best saved weights for ResNet18
resnet_model.load_state_dict(
    torch.load('../results/ResNet_best.pth',
               map_location=device)
)

# Get predictions
resnet_labels, resnet_preds = get_predictions(resnet_model, test_loader)

# Calculate accuracy
resnet_test_acc = (resnet_labels == resnet_preds).mean() * 100

print(f"ResNet18 Transfer — Test Accuracy: {resnet_test_acc:.2f}%")
print("=" * 55)
print(classification_report(resnet_labels, resnet_preds, 
                             target_names=CLASS_NAMES))
```

```python
def plot_confusion_matrix(labels, preds, model_name):
    cm = confusion_matrix(labels, preds)

    plt.figure(figsize=(7, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=CLASS_NAMES,
                yticklabels=CLASS_NAMES,
                linewidths=0.5)
    plt.title(f'{model_name} — Confusion Matrix', 
              fontweight='bold', fontsize=13)
    plt.ylabel('True Label',      fontsize=11)
    plt.xlabel('Predicted Label', fontsize=11)
    plt.tight_layout()
    plt.savefig(f'../results/{model_name.lower().replace(" ","_")}_confusion.png',
                dpi=150, bbox_inches='tight')
    plt.show()
    print(f"Saved!")

# Plot confusion matrix for both models
plot_confusion_matrix(cnn_labels,    cnn_preds,    'Custom CNN')
plot_confusion_matrix(resnet_labels, resnet_preds, 'ResNet18 Transfer')
```

## Step 10 — Visualise Sample Predictions

```python
def visualise_predictions(model, model_name, loader, n=8):
    """Show n test images with predicted vs true labels."""
    model.eval()
    imgs_list   = []
    preds_list  = []
    labels_list = []

    with torch.no_grad():
        for images, labels in loader:
            outputs = model(images.to(device))
            _, preds = outputs.max(1)
            imgs_list.extend(images.cpu())
            preds_list.extend(preds.cpu().numpy())
            labels_list.extend(labels.numpy())
            if len(imgs_list) >= n:
                break

    # Denormalise images for display
    mean = np.array([0.485, 0.456, 0.406])
    std  = np.array([0.229, 0.224, 0.225])

    fig, axes = plt.subplots(2, 4, figsize=(15, 8))
    fig.suptitle(f'{model_name} — Sample Predictions',
                 fontsize=13, fontweight='bold')

    for i, ax in enumerate(axes.flat):
        # Denormalise
        img = imgs_list[i].permute(1, 2, 0).numpy()
        img = np.clip(img * std + mean, 0, 1)

        pred  = CLASS_NAMES[preds_list[i]]
        true  = CLASS_NAMES[labels_list[i]]

        # Green = correct, Red = wrong
        color = 'green' if preds_list[i] == labels_list[i] else 'red'

        ax.imshow(img)
        ax.set_title(f'Pred: {pred}\nTrue: {true}',
                     color=color, fontsize=9, fontweight='bold')
        ax.axis('off')

    plt.tight_layout()
    plt.savefig(f'../results/{model_name.lower().replace(" ","_")}_predictions.png',
                dpi=150, bbox_inches='tight')
    plt.show()
    print(f"Saved!")


# Visualise predictions for both models
visualise_predictions(cnn_model,    'Custom CNN',        test_loader)
visualise_predictions(resnet_model, 'ResNet18 Transfer', test_loader)
```

```python
from PIL import Image
import torchvision.transforms as transforms
import torch
import matplotlib.pyplot as plt

def predict_single_image(image_path):
    """
    Give it any brain MRI image path
    It will predict the tumor type!
    """

    # ── Load and display the image ────────────────────
    img = Image.open(image_path).convert('RGB')
    
    plt.figure(figsize=(4, 4))
    plt.imshow(img, cmap='gray')
    plt.title('Input MRI Image', fontweight='bold')
    plt.axis('off')
    plt.show()

    # ── Preprocess the image ──────────────────────────
    transform = transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.ToTensor(),
        transforms.Normalize(
            mean=[0.485, 0.456, 0.406],
            std=[0.229, 0.224, 0.225]
        )
    ])

    # Add batch dimension: (3, 224, 224) → (1, 3, 224, 224)
    img_tensor = transform(img).unsqueeze(0).to(device)

    # ── Predict using ResNet18 ────────────────────────
    resnet_model.eval()
    with torch.no_grad():
        outputs     = resnet_model(img_tensor)
        
        # Get probabilities using softmax
        probs       = torch.softmax(outputs, dim=1)[0]
        
        # Get the predicted class
        pred_idx    = probs.argmax().item()
        pred_class  = CLASS_NAMES[pred_idx]
        confidence  = probs[pred_idx].item() * 100

    # ── Show results ──────────────────────────────────
    print("=" * 40)
    print(f"  PREDICTION: {pred_class}")
    print(f"  CONFIDENCE: {confidence:.2f}%")
    print("=" * 40)

    # Show all class probabilities
    print("\nAll probabilities:")
    for i, (cls, prob) in enumerate(zip(CLASS_NAMES, probs)):
        bar    = "█" * int(prob.item() * 30)
        marker = " ← predicted" if i == pred_idx else ""
        print(f"  {cls:<12} {prob.item()*100:5.1f}%  {bar}{marker}")

    return pred_class, confidence
```

```python
# ── Fix: define device and CLASS_NAMES ───────────────
import torch
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
CLASS_NAMES = ['Glioma', 'Meningioma', 'No Tumor', 'Pituitary']
```
